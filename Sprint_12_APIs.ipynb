### 1.1 Consultas GET básicas

Consulta l’API pública JSONPlaceholder usando el método GET para obtener:

- Llista de publicacions (/posts)  
- Llista d’usuaris (/users)  
- Llista de tasques (/todos)  

### 1.2 Muestra cantidad de cada recurso y estado de la petición

```python
# Lista de posts
posts_resp = requests.get(url_nivel_1 + 'posts')
print('Posts - Codigo estado:', posts_resp.status_code)
print('Total posts:', len(posts_resp.json()))

# Llista d'usuaris
users_resp = requests.get(url_nivel_1 + 'users')
print('\nUsuaris - Codigo estado:', users_resp.status_code)
print('Total usuaris:', len(users_resp.json()))

# Llista de tasques
todos_resp = requests.get(url_nivel_1 + 'todos')
print('\nTodos - Codigo estado:', todos_resp.status_code)
print('Total todos:', len(todos_resp.json()))
```

### 1.3. Obten Error 404

```python
# Publicación con url no válida
error_resp = requests.get(url_nivel_1 + 'posts/url_invalida')
print('Respuesta a consulta errónea:', error_resp.status_code)
```

### 1.4. POST nueva publicación
Haz una petición `POST` para crear una nueva publicación fictícia. Incluye el título, el cuerpo del mensaje y un user Id.

```python
nova_post = {'title': 'NOU TÍTOL VAN', 'body': 'Aquí va el text de la nova publicació de la sra. Van', 'userId': 10}
post_resp = requests.post(url_nivel_1 + 'posts', json=nova_post)
print('POST - Codi estat:', post_resp.status_code)
print('Resposta JSON:', post_resp.json())
```

### 1.5. PATCH modificar

```python
patch_data = {'title': 'El meu Títol modificat'}
patch_resp = requests.patch(url_nivel_1 + 'posts/1', json=patch_data)
print('PUT - Codi estat:', patch_resp.status_code)
print('Resposta JSON:', patch_resp.json())
```

### 1.6. DELETE

```python
delete_resp = requests.delete(url_nivel_1 + 'posts/1')
print('DELETE - Codi estat:', delete_resp.status_code)
print('Resposta JSON:', delete_resp.json())
```

```python
for flight in api_response['data']:
        airline = flight['airline']['name'] if flight['airline'] else 'Desconeguda'
        flight_iata = flight['flight']['iata'] if flight['flight'] else 'N/A'
        dep_airport = flight['departure']['airport'] if flight['departure'] else 'N/A'
        dep_iata = flight['departure']['iata'] if flight['departure'] else 'N/A'
        arr_airport = flight['arrival']['airport'] if flight['arrival'] else 'N/A'
        arr_iata = flight['arrival']['iata'] if flight['arrival'] else 'N/A'
        print(airline, flight_iata, dep_airport,arr_airport, arr_iata)
```

#### Buscar un dataset de tu interés mediante package_search y package_show.

```python
paraules_clau = ["transit aeri"]
base_url = "https://opendata-ajuntament.barcelona.cat/data/api/3/action"


resultats_totals = []

for paraula in paraules_clau:
    url = f"{base_url}/package_search"
    params = {"q": paraula, "rows": 10}
    response = requests.get(url, params=params)
    data = response.json()
    
    resultats = data["result"]["results"]
    print(f"\n '{paraula}': {len(resultats)} resultats:")
    for resultat in resultats:
        print(f"  - {resultat['name']} | {resultat['title']}")
        resultats_totals.append({
            "paraula_clau": paraula,
            "name": resultat["name"],
            "title": resultat["title"],
            "num_resources": len(resultat.get("resources", []))
            })

print(f"\n Total resultats trobats: {len(resultats_totals)}")
```
